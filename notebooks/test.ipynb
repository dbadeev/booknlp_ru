```python
from pathlib import Path
import os

# Получаем путь к корню проекта (родитель текущей папки notebooks)
# os.getcwd() вернет .../booknlp_ru/notebooks
# .parent вернет .../booknlp_ru
project_root = Path(os.getcwd()).parent

# Строим правильный путь
raw_path = project_root / "data" / "raw"

print(f"Ищем файлы в: {raw_path}\n")

files = list(raw_path.rglob("*.conllu"))

if not files:
    print("❌ Файлы всё ещё не найдены. Проверьте структуру папок.")
else:
    print(f"✅ Найдено {len(files)} файлов:")
    for f in files:
        print(f" - {f.name}")
```
